## Plots for DataFrames or Series

> Working with datasets on a daily basis has made it easier for me to read and understand table statistics. However, while numeric statistics might give you the essence of your data, a graph or visualization can uncover a whole new dimension of underlying information within your dataset.

When visualizing data, it is important to understand 3 things.
- What are the different types of plots you can use?
- How many should you use and how would you explain them?
- Can you tell a story using just these plots? What do they tell you?

In the next 3 cells we will prepare the DataFrames that we will be using for plotting

```python
# prepare the data for plotting
# Import required library
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline

mpl.rcParams['font.size'] = 12
mpl.rcParams['figure.dpi'] = 90.

# we will rename our DataFrame to cases
cases = df.copy()
daily_cases = df.diff(axis=1)  # this DF contains the daily new cases

# import the number of deaths
file = 'data/deaths.csv'
deaths = pd.read_csv(file, index_col=0)
deaths.index.name = None

daily_deaths = deaths.diff(axis=1)  # this contains the daily number of deaths
```

When comparing countries (i.e. US with Sweden) one might instead compare the number per inhabitants:

```python
# DataFrame containing populations of all countries
population_2018 = pd.read_csv('data/population_2018.csv',squeeze=True)
# population_2018.sort_values(['Value'], ascending=False, axis=0, inplace=True)
population_2018.set_index(['Unnamed: 0'], inplace=True)
population_2018.index.name = None
```

```python
# normalise the cases to cases per 100k inhabitants
# transform to a series
population_2018 = population_2018['Value']
cases_pI = cases.div(population_2018.loc[cases.index],axis=0) * 100000
deaths_pI = deaths.div(population_2018.loc[deaths.index],axis=0) * 100000

daily_cases_pI = cases_pI.diff(axis=1)
daily_deaths_pI = deaths_pI.diff(axis=1)

last_day = cases_pI.columns[-1]
```

So we have created 8 DataFrames that we can use for plotting (<code>cases</code>, <code>daily_cases</code>, <code>deaths</code>, <code>daily_deaths</code> and 4 more normalized to inhabitants <code>..._pI</code> ).

### Scatter plots

Scatter plots are great and can be used to describe relationships, look at individual sections of data, and describe the distribution of data.
What relation is between the daily new infections and daily new deaths? Can we find some structure or patterns?

```python
# daily is a DataFrame containing two columns, daily_cases and daily_deaths in country
country = 'Italy'

daily = daily_cases.loc[[country]].transpose()
daily.rename(columns={country: 'daily_cases'}, inplace=True)
daily['daily_deaths'] = daily_deaths.loc[country]
daily.head()
```

```python
# this creates a scatter plot
daily.plot(kind='scatter', x='daily_cases', y='daily_deaths');
```

You can clearly see a linear trend, however, if you look closer you see that at daily_cases$\approx$1000 the slop is changing.

```python
# we will try to apply a linear regression to the two time periods

days_first = 45  # number of days of the first period
# for italy you might use 45, Sweden around 60

ax = daily.iloc[:days_first].plot(kind='scatter',
                                  x='daily_cases',
                                  y='daily_deaths',
                                  color='red')

daily.iloc[days_first:].plot(kind='scatter',
                             x='daily_cases',
                             y='daily_deaths',
                             color='blue',
                             ax=ax)

# lets do a linear regression
from sklearn import linear_model

regr = linear_model.LinearRegression()
x = daily.iloc[:days_first][['daily_cases']].fillna(0)
y = daily.iloc[:days_first][['daily_deaths']].fillna(0)

regr.fit(x, y)
# The coefficients
print('Coefficient: ', regr.coef_)

x = daily_cases.loc[[country]].transpose()[:].fillna(0)
y = regr.predict(x)
ax.plot(x, y)

x = daily.iloc[days_first:][['daily_cases']].fillna(0)
y = daily.iloc[days_first:][['daily_deaths']].fillna(0)
regr.fit(x, y)
print('Coefficient: ', regr.coef_)
x = daily_cases.loc[[country]].transpose()[:].fillna(0)
y = regr.predict(x)
ax.plot(x, y);
```

```python
ax = daily[['daily_cases', 'daily_deaths']].plot(figsize=(8, 6))
daily[['daily_cases','daily_deaths']].rolling(7).mean().plot(ax=ax);
```

Okay, still we can add more information to the scatter plot, what about the order in time:

### Italy data COVID-19

This is a nice example of how to use the Googletrans API:

```python
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates

url = 'https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-andamento-nazionale/'
df = pd.read_csv(url + 'dpc-covid19-ita-andamento-nazionale.csv')

df.head()
```

The columns are in Italian. Let's try to translate this directly in Python. DeepL is only for paying subscribers, so Google Translate will be used.

> Googletrans is a free and unlimited python library that implemented Google Translate API. This uses the Google Translate Ajax API to make calls to such methods as detect and translate.

```python
# in case you have not installed it yet, uncomment the next line of code
!pip install googletrans
from googletrans import Translator

# get languanges with 'googletrans.LANGCODES'
col_list = []
for col in df.columns:
    col_list.append(col.replace('_', ' '))

english_cols = Translator().translate(col_list, src='it', dest='en')

for i in range(len(english_cols)):
    english_cols[i] = english_cols[i].text

df.columns = english_cols
date_format = [pd.to_datetime(d) for d in df['date']]
df['date'] = date_format
df['day_of_week'] = df['date'].dt.dayofweek
df = df.drop('date', 1)
df.index = date_format
df.head()
```

Well, now we have translated the columns from italian to english and we have changed the date in the csv file to a proper DateTime and replaced the Index of the DataFrame with the Date. Furthermore, we have added a new columns 'day_of_week'. Now you can try to continue.

It is known that regions of Lombardia, Piemonte, Veneto and Emilia-Romagna have been affected more severely than the others. In order to quantify this, we can inspect the regional COVID-19 dataset provided by the Civil Protection, and calculate the proportion of the deaths registered in those regions (we call them Zone 1)

```python
data = pd.read_csv(
    "https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-regioni/dpc-covid19-ita-regioni.csv"
)

zone1_df = data[data.denominazione_regione.isin(
    ['Piemonte', 'Emilia-Romagna', 'Veneto', 'Lombardia'])]

zone1_df['deceduti'].sum()

print("Zone 1 accounts for %s percent of the total deaths" %
      (round(zone1_df['deceduti'].sum() / data['deceduti'].sum() * 100, 2)))
```
